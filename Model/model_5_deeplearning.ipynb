```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
#X = df.drop("koi_disposition", axis=1)
X_selected=df[["koi_fpflag_nt","koi_fpflag_ss","koi_fpflag_co","koi_fpflag_ec",
              "koi_period","koi_period_err1","koi_period_err2",
              "koi_time0bk","koi_time0bk_err1","koi_time0bk_err2",
              "koi_duration","koi_duration_err1","koi_duration_err2",
               "koi_depth","koi_depth_err1","koi_depth_err2",
              "koi_model_snr","koi_steff","koi_steff_err1","koi_steff_err2"]]
```

```python
len(X_selected.columns)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df["koi_disposition"]
print(X_selected.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, random_state=1)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
label_encoder.inverse_transform([0,1,2])
```

# Train the Model



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create model and add layers
model = Sequential()
model.add(Dense(units=60, activation='relu', input_dim=20))
model.add(Dense(units=60, activation='relu'))
model.add(Dense(units=3, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train_scaled, y_train_categorical, epochs=100, shuffle=True, verbose=2)
```

```python
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
def create_model(optimizer='adam'):
    model = Sequential()
    model.add(Dense(units=60, activation='relu', input_dim=20))
    model.add(Dense(units=60, activation='relu'))
    model.add(Dense(units=3, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy']) 
    return model
```

```python
# Create the GridSearchCV model
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

model = KerasClassifier(build_fn=create_model)
optimizers = ['adam','rmsprop']
epochs = np.array([50, 100, 200])

param_grid = {'optimizer':optimizers,
              'epochs':[50,100,200,500]}

grid = GridSearchCV(model, param_grid, verbose=3, n_jobs=-1)
```
